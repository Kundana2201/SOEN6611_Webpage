#### This block of the script reads Code Complexity from the CSV file extracted from JUnit results of seven different versions of Apache Commons IO projects.

#### This block of code calculates and prints the value of spearman correlation between Statement coverage and Code complexity. And, spearman correlation between Branch coverage and Code complexity. The value should be in the range [-1, 1] 
#### The results indicates that there exists a strong positive correlation between both "(Statement coverage, Code complexity)", and between "(Branch coverage and Code complexity)".

```python
from scipy.stats import spearmanr
print(spearmanr(state, com)) #Correlation between Statement Coverage & Complexity
print(spearmanr(branch, com)) #Correlation between Branch Coverage & Complexity
```

#### Correlation between Statement coverage and Code complexity is represented using a scatter plot below.

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy, pylab
import plotly.plotly as py
import plotly.tools as tls
plt.scatter(state, com)
plt.xlim(0, 7000)
plt.ylim(0, 300)
plt.yticks(np.arange(0, 300, 50))
plt.xticks(np.arange(0, 7000, 1000))
plt.ylabel("Statement Coverage")
plt.xlabel("Code Complexity")
plt.title("Statement Coverage vs Code Coverage")
plt.show()
```
