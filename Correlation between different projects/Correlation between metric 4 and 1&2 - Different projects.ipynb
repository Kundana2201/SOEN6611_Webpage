#### This block of the script reads the Statement Coverage from the CSV file extracted from JUnit results of five different projects.
#### Statement Coverage results are stored in column 4 of the CSV file.
#### for loop and if condition at the end of the block is used to read only the numeric values from the mentioned column.

```python
import csv
state = []
with open("C:\\Users\\arvin\\Desktop\\csv\\collections.csv") as csvDataFile:
    statement_coverage1=list(csv.reader(csvDataFile))
with open("C:\\Users\\arvin\\Desktop\\csv\\Config.csv") as csvDataFile:
    statement_coverage1.extend(list(csv.reader(csvDataFile)))
with open("C:\\Users\\arvin\\Desktop\\csv\\fu.csv") as csvDataFile:
    statement_coverage1.extend(list(csv.reader(csvDataFile)))
with open("C:\\Users\\arvin\\Desktop\\csv\\io.csv") as csvDataFile:
    statement_coverage1.extend(list(csv.reader(csvDataFile)))
with open("C:\\Users\\arvin\\Desktop\\csv\\jfreechart-master.csv") as csvDataFile:
    statement_coverage1.extend(list(csv.reader(csvDataFile)))
    count=1
    for column in statement_coverage1:
        if(column[4].isdigit()):
            state.append(int(column[4]))
        else:
            continue
print("Class wise Statement Coverage for different projects\n")
print(state)
```

#### This block of the script reads Code Complexity from the CSV file extracted from JUnit results of five different projects.
#### Code Complexity results are stored in column 10 of the CSV file.
#### for loop and if condition at the end of the block is used to read only the numeric values from the mentioned column.

```python
import csv
com = []
with open("C:\\Users\\arvin\\Desktop\\csv\\collections.csv") as csvDataFile:
    code_complexity1=list(csv.reader(csvDataFile))
with open("C:\\Users\\arvin\\Desktop\\csv\\Config.csv") as csvDataFile:
    code_complexity1.extend(list(csv.reader(csvDataFile)))
with open("C:\\Users\\arvin\\Desktop\\csv\\fu.csv") as csvDataFile:
    code_complexity1.extend(list(csv.reader(csvDataFile)))
with open("C:\\Users\\arvin\\Desktop\\csv\\io.csv") as csvDataFile:
    code_complexity1.extend(list(csv.reader(csvDataFile)))
with open("C:\\Users\\arvin\\Desktop\\csv\\jfreechart-master.csv") as csvDataFile:
    code_complexity1.extend(list(csv.reader(csvDataFile)))
    count=1
    for column in branch_coverage1:
        if(column[10].isdigit()):
            com.append(int(column[10]))
        else:
            continue
print("Class wise Code Complexity for different projects\n")
print(com)
```

#### This block of code calculates and prints the value of spearman correlation between Code complexity and Statement coverage. And, spearman correlation between Code complexity and Branch coverage. The value should be in the range [-1, 1] 
#### The results indicatres that there exists a strong positive correlation between both "(Code complexity, Statement coverage)", and between "(Code complexity and Branch coverage)".

```python
from scipy.stats import spearmanr
print(spearmanr(com, state)) #Correlation between Code Complexity and Statement Coverage 
print(spearmanr(com, branch)) #Correlation between Code Complexity and Branch Coverage
```

#### Correlation between Code complexity and Statement coverage is represented using a scatter plot below.

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy, pylab
import plotly.plotly as py
import plotly.tools as tls
plt.scatter(state, com)
plt.xlim(0, 7000)
plt.ylim(0, 300)
plt.yticks(np.arange(0, 300, 50))
plt.xticks(np.arange(0, 7000, 1000))
plt.ylabel("Statement Coverage")
plt.xlabel("Code Complexity")
plt.title("Code Complexity Vs Statement Coverage")
plt.show()
```
